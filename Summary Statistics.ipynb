## Summary statistics on land use & percentage of use

```python
land_use = pd.read_csv('./land_use.csv')
```

```python
land_use.count()
```

```python
land_use = land_use[['area_ha','zonecode1']]
```

```python
land_use.head()
```

```python
land_use_grouped_sum = land_use.groupby(['zonecode1'])['area_ha'].sum()
```

```python
land_use_grouped_sum = land_use_grouped_sum.reset_index()
# land_use_grouped_sum
```

```python
land_use_grouped_sum['area_sum'] = land_use_grouped_sum['area_ha'].sum()
```

```python
land_use_grouped_sum['percentage_of_area'] = (land_use_grouped_sum['area_ha']/land_use_grouped_sum['area_sum'])*100
```

```python
land_use_grouped_sum.columns = ['land_use','sum_area','all_sum','percentage_area']
```

```python
# land_use_grouped_sum
```

```python
land_use_grouped_p = land_use_grouped_sum.copy()
```

```python
land_use_grouped_p = land_use_grouped_p.drop(['all_sum'], axis=1)
```

```python
land_use_grouped_p = land_use_grouped_p.drop(['sum_area'], axis=1)
```

```python
land_use_grouped_p
```

```python
land_use_grouped_sum = land_use_grouped_sum.drop(['percentage_area'], axis=1)
land_use_grouped_sum = land_use_grouped_sum.drop(['all_sum'], axis=1)
```

```python
# land_use_grouped_sum
```

```python
land_use_grouped_mean = land_use.groupby(['zonecode1'])['area_ha'].mean()
land_use_grouped_mean = land_use_grouped_mean.reset_index()
land_use_grouped_mean.columns = ['land_use','mean_area']
```

```python
land_use_grouped_median = land_use.groupby(['zonecode1'])['area_ha'].median()
land_use_grouped_median = land_use_grouped_median.reset_index()
land_use_grouped_median.columns = ['land_use','median_area']
```

```python
land_use_grouped_min = land_use.groupby(['zonecode1'])['area_ha'].min()
land_use_grouped_min = land_use_grouped_min.reset_index()
land_use_grouped_min.columns = ['land_use','min_area']
```

```python
land_use_grouped_max = land_use.groupby(['zonecode1'])['area_ha'].max()
land_use_grouped_max = land_use_grouped_max.reset_index()
land_use_grouped_max.columns = ['land_use','max_area']
```

```python
land_use_grouped_stddev = land_use.groupby(['zonecode1'])['area_ha'].std()
land_use_grouped_stddev = land_use_grouped_stddev.reset_index()
land_use_grouped_stddev.columns = ['land_use','std_area']
```

```python
land_use_grouped_count = land_use.groupby(['zonecode1'])['area_ha'].count()
land_use_grouped_count = land_use_grouped_count.reset_index()
land_use_grouped_count.columns = ['land_use','count']
```

```python
land_use_grouped_mean
```

```python
land_use_grouped_median
```

```python
land_use_grouped_min
```

```python
land_use_grouped_max
```

```python
land_use_grouped_stddev
```

```python
land_use_summary = pd.merge(land_use_grouped_sum, land_use_grouped_mean, on='land_use', how='outer')
```

```python
land_use_summary = pd.merge(land_use_summary, land_use_grouped_median, on='land_use', how='outer')
```

```python
land_use_summary = pd.merge(land_use_summary, land_use_grouped_min, on='land_use', how='outer')
```

```python
land_use_summary = pd.merge(land_use_summary, land_use_grouped_max, on='land_use', how='outer')
```

```python
land_use_summary = pd.merge(land_use_summary, land_use_grouped_stddev, on='land_use', how='outer')
```

```python
land_use_summary = pd.merge(land_use_summary, land_use_grouped_p, on='land_use', how='outer')
```
