K-means

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Importing the dataset
data = pd.read_csv('./merge_5_final.csv')
```

```python
data.head()
```

```python
data = data[['area_ind','area_trans','area_res','area_inst','area_com','area_nat','area_os']]
```

```python
data.head()
```

```python
# Convert DataFrame to matrix
mat = data.values
# Using sklearn
km2 = KMeans(n_clusters=2).fit(mat)
# Get cluster assignment labels
labels2 = km2.labels_
# Format results as a DataFrame
#results = pd.DataFrame([data.index,labels]).T
km3 = KMeans(n_clusters=3).fit(mat)
labels3 = km3.labels_
# 4 clusters
km4 = KMeans(n_clusters=4).fit(mat)
labels4 = km4.labels_
```

```python
data['k_means_2']=labels2
data['k_means_3']=labels3
data['k_means_4']=labels4
```

```python
data
```

```python
data = data.reset_index()
```

```python
data = data[['index','k_means_2','k_means_3','k_means_4']]
```

```python
data.head()
```

```python
data2 = pd.read_csv('./merge_5_final.csv')
```

```python
data3 = pd.merge(data, data2, on='index', how='outer')
```
